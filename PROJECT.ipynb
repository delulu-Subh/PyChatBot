````python
import openai
import os
from dotenv import load_dotenv, find_dotenv

# Load environment variables from the .env file
_ = load_dotenv(find_dotenv())

# Get the API key from the environment variable named 'SAMBANOVA_API_KEY'
api_key = os.getenv('SAMBANOVA_API_KEY')

# Optional: Add a check to ensure the API key was loaded
if api_key is None:
    raise ValueError("SambaNova API Key not found. Make sure SAMBANOVA_API_KEY is set in your .env file.")

# Initialize the OpenAI client with the SambaNova API key and, crucially, the base_url
client = openai.OpenAI(
    api_key=api_key,
    base_url="https://api.sambanova.ai/v1"
)

print("OpenAI client initialized with SambaNova base URL.")

def get_completion(prompt, model="Meta-Llama-3.1-8B-Instruct", temperature=0):
    messages = [{"role": "user", "content": prompt}]
    try:
        response = client.chat.completions.create(
            model=model,
            messages=messages,
            temperature=temperature,
        )
        return response.choices[0].message.content
    except openai.APIStatusError as e:
        print(f"API Error ({model}): {e.status_code} - {e.response}")
        print(f"Response data: {e.body}")
        return f"Error: Failed to get completion - {e.status_code}"
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return f"Error: An unexpected error occurred: {e}"

# --- End of Core Setup ---
# All subsequent tasks (Summarizing, Inferring, Transforming, Expanding, Chatbot)
# will use the 'get_completion' function defined above.
````

````python
## 1. Summarizing Text
print("--- Summarization Examples ---")

# Example 1: Summarize into a specific number of sentences
long_text_1 = """
The Amazon rainforest is the largest rainforest in the world, covering an area of approximately 5.5 million square kilometers.
It spans nine countries, with the majority of it located in Brazil.
The rainforest is incredibly biodiverse, home to an estimated 10% of the world's known species, including countless insects, plants, birds, and mammals.
It plays a crucial role in regulating the Earth's climate by absorbing vast amounts of carbon dioxide and releasing oxygen.
Deforestation, primarily due to agricultural expansion and logging, poses a significant threat to the Amazon's ecosystem and its indigenous communities.
Conservation efforts are ongoing to protect this vital natural resource.
"""
prompt_summary_sentences = f"""
Summarize the following text into exactly two sentences.
Text: ```{long_text_1}```
"""
response_summary_sentences = get_completion(prompt_summary_sentences)
print(f"**Summary (2 sentences):**\n{response_summary_sentences}\n")

# Example 2: Summarize into bullet points focusing on key aspects
long_text_2 = """
A recent study published in Nature magazine revealed new insights into the migratory patterns of monarch butterflies.
Researchers tracked thousands of butterflies using miniature GPS devices, discovering previously unknown resting points and flight corridors.
The study highlights the critical importance of preserving diverse habitats along their migratory routes, particularly milkweed patches which are essential for their larvae.
Climate change and pesticide use are identified as major threats contributing to the decline in monarch populations.
"""
prompt_summary_bullets = f"""
Summarize the following study findings into a bulleted list, focusing on discoveries, importance, and threats.
Text: ```{long_text_2}```
"""
response_summary_bullets = get_completion(prompt_summary_bullets)
print(f"**Summary (Bullet Points):**\n{response_summary_bullets}\n")

# Example 3: Summarize a customer review for conciseness
customer_review = """
I bought this new noise-cancelling headphones last week, and I must say I'm thoroughly impressed with the sound quality.
The noise cancellation feature is truly exceptional; I could barely hear anything during my commute on a busy train.
However, the battery life is a bit shorter than advertised, I only got about 15 hours instead of 20, which is a minor inconvenience.
Also, the earcups are a bit too small for my ears, causing some discomfort after extended use, but overall, it's a great product.
"""
prompt_review_summary = f"""
Summarize the key pros and cons from the following customer review.
Text: ```{customer_review}```
"""
response_review_summary = get_completion(prompt_review_summary)
print(f"**Customer Review Summary:**\n{response_review_summary}\n")
````

````python
## 2. Inferring Information
print("--- Inferring Information Examples ---")

# Example 1: Sentiment Analysis
review_sentiment = """
I absolutely loved this movie! The plot was engaging, and the acting was superb.
I can't wait to see it again.
"""
prompt_sentiment = f"""
What is the sentiment of the following movie review?
Provide your answer as "positive", "negative", or "neutral".

Review: ```{review_sentiment}```
"""
response_sentiment = get_completion(prompt_sentiment)
print(f"**Sentiment Analysis:** {response_sentiment}\n")

# Example 2: Topic Extraction
news_article = """
Scientists at the Mars Rover mission announced a significant discovery today:
evidence of ancient liquid water on the Martian surface.
This finding strengthens the hypothesis that Mars may have once harbored life.
Further missions are planned to investigate subsurface ice deposits and atmospheric composition.
"""
prompt_topics = f"""
Identify the main topics discussed in the following news article.
List them as comma-separated keywords.

Article: ```{news_article}```
"""
response_topics = get_completion(prompt_topics)
print(f"**Topics:** {response_topics}\n")

# Example 3: Named Entity Recognition / Information Extraction (structured output)
customer_query = """
My order number is #XYZ-7890. I placed it on June 15, 2025.
The item is the "Quantum Smartwatch Pro" and it cost $299.99.
It hasn't arrived yet. My email is support@example.com.
"""
prompt_extract_info = f"""
Extract the following information from the customer query:
- Order Number
- Order Date
- Product Name
- Price
- Customer Email

Return the information in a JSON object.

Query: ```{customer_query}```
"""
response_extract_info = get_completion(prompt_extract_info)
print(f"**Extracted Information (JSON):**\n{response_extract_info}\n")

# Example 4: Identifying emotions
text_emotion = """
I'm so frustrated with the slow internet connection! It's impossible to get any work done.
"""
prompt_emotion = f"""
What emotion is the speaker expressing in the following text?
Text: ```{text_emotion}```
"""
response_emotion = get_completion(prompt_emotion)
print(f"**Expressed Emotion:** {response_emotion}\n")
````

````python
## 3. Transforming Text
print("--- Transforming Text Examples ---")

# Example 1: Language Translation
english_sentence = "The quick brown fox jumps over the lazy dog."
prompt_translate = f"""
Translate the following English text into Spanish.

Text: ```{english_sentence}```
"""
response_translate = get_completion(prompt_translate)
print(f"**Original (English):** {english_sentence}")
print(f"**Translation (Spanish):** {response_translate}\n")

# Example 2: Grammar and Spelling Correction
bad_text = "Thier going to the store, and its a good day for an walk."
prompt_grammar_check = f"""
Correct any grammar and spelling errors in the following text.

Text: ```{bad_text}```
"""
response_grammar_check = get_completion(prompt_grammar_check)
print(f"**Original:** {bad_text}")
print(f"**Corrected:** {response_grammar_check}\n")

# Example 3: Tone Adjustment (Formal to Casual)
formal_text = "We hereby confirm receipt of your esteemed inquiry regarding our product line."
prompt_casual_tone = f"""
Rewrite the following formal sentence in a more casual and friendly tone.

Text: ```{formal_text}```
"""
response_casual_tone = get_completion(prompt_casual_tone)
print(f"**Original (Formal):** {formal_text}")
print(f"**Rewritten (Casual):** {response_casual_tone}\n")

# Example 4: Formatting (Text to HTML)
list_items = """
- Item One
- Item Two
- Item Three
"""
prompt_to_html = f"""
Convert the following list into an HTML unordered list.

List: ```{list_items}```
"""
response_to_html = get_completion(prompt_to_html)
print(f"**Original List:**\n{list_items}")
print(f"**Converted to HTML:**\n{response_to_html}\n")
````

````python
## 4. Expanding Text
print("--- Expanding Text Examples ---")

# Example 1: Generate an email
email_context = {
    "sender": "Alice",
    "recipient": "Bob",
    "topic": "meeting reschedule",
    "old_time": "Monday at 10 AM",
    "new_time": "Wednesday at 2 PM",
    "reason": "unexpected client visit"
}
prompt_email_generation = f"""
Draft a polite email from {email_context['sender']} to {email_context['recipient']}
to reschedule a meeting.

Original meeting time: {email_context['old_time']}
New proposed time: {email_context['new_time']}
Reason for reschedule: {email_context['reason']}

Start the email with "Dear {email_context['recipient']},"
"""
# Use a slightly higher temperature for more varied output
response_email_generation = get_completion(prompt_email_generation, temperature=0.7)
print(f"**Generated Email:**\n{response_email_generation}\n")

# Example 2: Generate a product description
product_features = """
- Product: "Eco-Friendly Water Bottle"
- Material: Recycled plastic
- Capacity: 750ml
- Features: Double-walled insulation, leak-proof lid, available in multiple colors.
- Target Audience: Environmentally conscious individuals, fitness enthusiasts.
"""
prompt_product_description = f"""
Write a compelling and concise product description for the following product.
Highlight its key features and benefits for the target audience.

{product_features}
"""
response_product_description = get_completion(prompt_product_description, temperature=0.8)
print(f"**Generated Product Description:**\n{response_product_description}\n")

# Example 3: Generate a short marketing blurb
service_info = """
Service: "AI-Powered Personal Finance Advisor"
Benefits: Budgeting, investment advice, debt management, personalized insights.
Value Proposition: Helps users make smarter financial decisions effortlessly.
"""
prompt_marketing_blurb = f"""
Create a catchy, short marketing blurb (2-3 sentences) for the following service.

{service_info}
"""
response_marketing_blurb = get_completion(prompt_marketing_blurb, temperature=0.9)
print(f"**Generated Marketing Blurb:**\n{response_marketing_blurb}\n")
````

````python
## 5. Building a Custom Chatbot
print("--- Custom Chatbot Example ---")

# The 'messages' list will hold the conversation history.
# The system message defines the chatbot's persona.
conversation_history = [
    {"role": "system", "content": "You are a friendly and helpful AI assistant that answers questions about Python programming."},
]

def chat_with_model_interactive(user_message, history, model="Meta-Llama-3.1-8B-Instruct"):
    history.append({"role": "user", "content": user_message}) # Add user's message to history
    
    try:
        response = client.chat.completions.create(
            model=model,
            messages=history, # Pass the entire history for context
            temperature=0.7,  # Higher temperature for more conversational and varied responses
        )
        assistant_reply = response.choices[0].message.content
        history.append({"role": "assistant", "content": assistant_reply}) # Add assistant's reply to history
        return assistant_reply
    except openai.APIStatusError as e:
        print(f"API Error ({model}): {e.status_code} - {e.response}")
        print(f"Response data: {e.body}")
        return "Error: I'm having trouble responding right now. Please try again later."
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return "Error: An unexpected error occurred while processing your request."

# Start the interactive chat loop
print("Chatbot: Hello! How can I help you with Python today? (Type 'quit' to exit)")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
        print("Chatbot: Goodbye! Happy coding!")
        break
    
    bot_response = chat_with_model_interactive(user_input, conversation_history)
    print(f"Chatbot: {bot_response}")

# Optional: Print the full conversation history after quitting for review
# print("\n--- Full Conversation History ---")
# for msg in conversation_history:
#     print(f"{msg['role'].capitalize()}: {msg['content']}")
````
